```python
from experta import *
```

```python
class AnimalClass(Fact):
    """млекопитающее, пресмыкающееся, рыбы, птицы"""
    pass

class LivingEnvironment(Fact):
    """водная или земная"""
    pass

class Fetus(Fact):
    pass

class Surface(Fact):
    pass

class Scales(Fact):
    pass

class Grass(Fact):
    pass

class Hooves(Fact):
    pass

class Artiodactyl(Fact):
    pass

class Horns(Fact):
    pass

class Wool(Fact):
    pass

class Bouncy(Fact):
    pass

class Mleko(Fact):
    pass

class Presm(Fact):
    pass

class Size(Fact):
    pass

class Thickening_head(Fact):
    pass

class Length(Fact):
    pass

class Bird(Fact):
    pass

class Feathers(Fact):
    pass

class Claw(Fact):
    pass

class Beak(Fact):
    pass
```

```python
class Engine(KnowledgeEngine):
    @DefFacts()
    def _initial_action(self):
        yield AnimalClass(input('Введите класс животного (млекопитающее, пресмыкающееся, рыбы, птицы) '))
        
    @Rule(AnimalClass('млекопитающее'))
    def get_Mleco_fact(self):
        self.declare(Grass(input('Они едят траву (да/нет)? ')))
        self.declare(Artiodactyl(input('Они парнокопытные (да/нет)? ')))
        self.declare(Horns(input('У них есть рога (да/нет)? ')))
        self.declare(Wool(input('У них длинная шерсть (да/нет)? ')))
        self.declare(Size(input('У них небольшой размер (да/нет)? ')))
        self.declare(Bouncy(input('Они прыгучие (да/нет)? ')))

    @Rule(AND(Wool('да'), Size('да'), Bouncy('да'), Grass('да'), Horns('да'), Artiodactyl('да')))
    def get_goat(self):
        print('Это козел')
      
    @Rule(AND(Wool('нет'), Size('нет'), Bouncy('нет'), Grass('да'), Horns('да'), Artiodactyl('да')))
    def get_bull(self):
        print('Это бык')
        
        
    @Rule(AnimalClass('пресмыкающееся'))
    def get_Presm_fact(self):
        self.declare(Thickening_head(input('У них есть утолщение около головы (да/нет)? ')))
        self.declare(Length(input('У них средняя длина(да/нет)? ')))
        
    @Rule(AND(Thickening_head('да'), Length('да')))
    def get_snake(self):
        print('Это змея')
        
        
    @Rule(AnimalClass('птицы'))
    def get_Bird_fact(self):
        self.declare(Feathers(input('Они имеют перья (да/нет)? ')))
        self.declare(Claw(input('Они имеют длинные когти (да/нет)? ')))
        self.declare(Beak(input('Они имеют длинный клюв (да/нет)? ')))
        
    @Rule(AND(Feathers('да'), Claw('да'), Beak('да')))
    def get_stork(self):
        print('это аист')
        
        
    @Rule(AnimalClass('рыбы'))
    def set_fetus_fact(self):
        if input('У них долгое вынашивание плода (нет/да)? ') == 'да':
            self.declare(Fetus())
            
        if input('У них чешуя (нет/да)? ') == 'да':
            self.declare(Scales())
            
    @Rule(AND(Scales(), Fetus()))
    def get_fish_answer(self):
        print('Это рыба')
        
    
prod = Engine()
prod.reset()
prod.run()
```
